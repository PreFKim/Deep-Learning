Classification 데이터 준비 과정
======================

```python
#본 코드는 CPU처리가 더 빠름
dataset_dir = '/content/drive/MyDrive/newstudy/ocean_trash_classfication/train'
data_dir_images = dataset_dir+'/images'


f = open(dataset_dir+'/train.csv','r')
data = csv.reader(f)

print(type(data))
print(data)
'''
for line in data:
  print(line)
'''
data = list(data)
```

ocean_trash_classification
====================

```python

success=0
failed = []

for i in range(1,len(data)):
  if os.path.exists(data_dir_images + '/'+data[i][0]):
    if os.path.exists(data_dir_images+'/'+data[i][1])== False:
      os.makedirs(data_dir_images+'/'+data[i][1])
    

    #파일의 정보표시 후 저장 (성공 개수/실행 횟수/총 개수) jpg로 저장시 이미지에 변화가 생김 ex) 0,0,255 -> 24,24,255
    print(f'분류 중:{data[i][0]}-({success}/{i}/'+str(len(data))+')')
    shutil.move(f'{data_dir_images}/{data[i][0]}',f'{data_dir_images}/{data[i][1]}/{data[i][0]}')
    success += 1
  else :
    failed.append(failed)

print(f'총 {len(data)}개 중 {success}개를 분류 하였습니다.')

if len(failed):
  print(f'{len(failed)}개의 파일 분류를 실패하였습니다.')
  print(failed)
```
